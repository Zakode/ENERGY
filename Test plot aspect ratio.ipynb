```python
import matplotlib.pyplot as plt
plt.plot([1,5],[2,3])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
axes=plt.gca()
axes.set_aspect(7,adjustable='datalim')
plt.show()
```

```python
fig,ax = plt.subplots()

plt.plot([1,5],[2,3])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

ax.set_aspect(5,adjustable='datalim')
plt.show()
```

```python
fig,ax = plt.subplots()

plt.plot([1,5],[2,3])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

ax.set_aspect('equal')
plt.show()
```
