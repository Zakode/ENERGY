## Load data and library to GCP

```python
%%bash 

function createStruct() {
  if [[ -d data && -d log && outputs ]] 
      then 
          echo "Dir already exist :"
      else     
          mkdir data log outputs tmp
          touch log/loadData.log log/instLib.log log/main.log
  fi
}


function getData() {

  if [[ -d data && -f "data/EdStatsCountry.csv" ]] 
      then 
          echo "EdStats files already in data/ :"
      else 
          echo "Downloading EdStats data/"
          wget -q -o log/loadData.log -P tmp "https://s3-eu-west-1.amazonaws.com/static.oc-static.com/prod/courses/files/Parcours_data_scientist/Projet+-+Donn%C3%A9es+%C3%A9ducatives/Projet+Python_Dataset_Edstats_csv.zip"
          unzip -d data tmp/Projet+Python_Dataset_Edstats_csv.zip >> log/loadData.log
  fi

  if [[ -d data && -f "data/countries_codes_and_coordinates.csv" ]] 
      then 
          echo "countries_coord file already in data/ :"
      else 
          wget -q -o log/loadData.log -P data "https://gist.githubusercontent.com/tadast/8827699/raw/f5cac3d42d16b78348610fc4ec301e9234f82821/countries_codes_and_coordinates.csv"
          cleanExt
  fi

  if [[ -d data && -f "data/california_islands.shp" ]] 
      then 
          echo "channelisland shapefil already in data/ :"
      else 
          wget -q -o log/loadData.log -P tmp http://stacks.stanford.edu/file/druid:nd958xt7750/data.zip
          unzip -d tmp/ tmp/data.zip
          mv tmp/california_islands.shp data/
          mv tmp/california_islands.shx data/
  fi

}



function cleanExt () {
  #clean data to simplify countries' names
  sed -i 's/"//g'  "data/countries_codes_and_coordinates.csv" 
  sed -i "s/, Plurinational State of//" "data/countries_codes_and_coordinates.csv" 
  sed -i "s/, the Democratic Republic of the//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Islamic Republic of//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/Korea, Democratic People's Republic of,/Korea N,/" "data/countries_codes_and_coordinates.csv"
  sed -i "s/Korea, Republic of, /Korea S,/" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, the former Yugoslav Republic of//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Federated States of//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Republic of//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Province of China//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, United Republic of//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Bolivarian Republic of//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Occupied//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, Ascension and Tristan da Cunha//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, British//" "data/countries_codes_and_coordinates.csv"
  sed -i "s/, U.S.//" "data/countries_codes_and_coordinates.csv"
}


###############################################
function main() {
  getData 
}

createStruct
main > log/main.log
```

```python
# for future usage check all col fill rate , unique value rate and duplicates
def clean1(df_raw) :
  """
  for a given dataframe, on a copy :
  - Drops empty variables
  - Strip categorial variables
  
  Returns the dataframe and a description dictionary  
  """

  df = df_raw.copy()
  nb_num = 0
  nb_cat = 0

  #print(f"""Sample Data BEFORE : """)
  #display(df.head(3))
  
  newRow = {}
  df_desc = {} #keys=["dataType","nbFilled","fillRate%","uniqueVal","uvRate%"])
  errors={}

  display(HTML(f"""
<i><strong>  Cleaining df : streaping columns and replacing white spaces in titles </i></strong>
  """))
  df.columns = df.columns.str.strip()
  df.columns = df.columns.str.replace(' ', '_')

  for col in df : 

    dataType= df[col].dtype
    nbFilled = df[col].notna().sum()
    fillRate = 100*nbFilled / len(df[col])
    uniqueVal = df[col].nunique()
    uvRate = 100*uniqueVal /  len(df[col])
    if fillRate == 0 :
      display(HTML(f"""<font color='red'>
      Droping empty column : {col} 
      </font>"""))
      df.drop([col], axis=1,inplace=True)
      #desc[col] = "empty variable dropped"
    else :
      #print(f"""
      #Checking Column : {col} 
      #""")
      newRow = {"dataType" : dataType,
                "nbFilled":nbFilled,
                "fillRate%":fillRate,
                "uniqueVal":uniqueVal,
                "uvRate%":uvRate
                }
      if pd.api.types.is_string_dtype(df[col] ) :
        try:
          df[col] = df[col].str.strip() #remove white spaces before and after 
        except Exception as e : 
          print(f"""
Error {e}
while stripping {col} to string           
          """)
          if_numeric = pd.to_numeric(df[col],errors='coerce')
          errors[col] = { "score numericability" : if_numeric.notna().sum()/len(if_numeric) ,
                         "series" : if_numeric}
          print(f"""
if {col} converted to numeric 
fillrate would be = {if_numeric.notna().sum()/len(if_numeric)}  
check errors for further treatment. 
{col} will be casted to string.    
          """)

          df[col] = df[col].apply(str).str.strip()
        finally :
          #print(f"{col} striped")
          nb_cat += 1
      else : 
        nb_num += 1
    
      df_desc[col] = newRow
 
  

  display(HTML(f"""
<h3>Sample Data AFTER Cleaning :</h3> 
<h4>
 <ul>
  <li>{df.shape[0]} individuals</li>
  <li>{df.shape[1]} variables : {nb_num} numeric / {nb_cat} categorial</li>
</ul>
</h4> 
  """))
  display(df.head(3))

  
  msno.matrix(df)
  plt.title("MISSING VALUES",fontdict={'color':  'darkblue',
                                     'weight': 'normal',
                                     'size': 32,
                                     })
  plt.show()
  
  
  return df,df_desc,errors
```

### Function describeNumCol

```python
def describeNumCol (df,colName) :
  """
  describe a numeric column with basic stats and graphics
  """

  dist_series = df[colName].value_counts()
  n = len(df[colName])

  #calculate sample kurtosis
  gamma1 = kurtosis(df[colName], bias=False)
  #The kurtosis of a normal distribution is 3.
  #kurtosis less than 3 : playkurtic, which means it tends to produce fewer and less extreme outliers than the normal distribution.
  #kurtosis greater than 3, : leptokurtic, which means it tends to produce more outliers than the normal distribution.
  if gamma1 <2.9 : 
    interp_gamma1 =  "playkurtic (concentrated)"
  elif gamma1 >3.1 :
    interp_gamma1 =  "leptokurtic (many outliers)"
  else : 
    interp_gamma1 = "distribution is normal"
  
  #calculate sample skewness
  gamma2 = skew(df[colName], bias=False)
  #A negative skew = the tail is on the left side of the distribution, which extends towards more negative values.
  #A positive skew = the tail is on the right side of the distribution, which extends towards more positive values.
  #A value of zero = there is no skewness in the distribution at all, meaning the distribution is perfectly symmetrical.
  if gamma2 <-0.2 : 
    interp_gamma2 =  "tail is on the left side of the distribution"
  elif gamma2 >0.2 :
    interp_gamma2 =  "tail is on the right side of the distribution"
  else : 
    interp_gamma2 = "distribution is symmetrical"


  #calculate Lorentz and Gini  
  lorenz = np.cumsum(np.sort(df[colName])) / df[colName].sum()
  lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
  AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
  S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
  gini = 2*S
  gini


  #prepare outputs
  graph = widgets.Output()
  table = widgets.Output(layout=widgets.Layout(margin='0 50px 0 0'))
  hLayout = widgets.HBox([table,graph])

  #sempre fare la bella figura
  fig,ax = plt.subplots(figsize=(20,15) )
  ax.xaxis.set_visible(False)
  ax.yaxis.set_visible(False)
  plt.title(f"Variable {colName}",
            fontdict={
                'fontsize': 24,
                'fontweight' : "bold" ,
                'verticalalignment': 'baseline',
                'horizontalalignment': 'center'},
            pad =50,
            )
  
  # grid specification
  gs = fig.add_gridspec(3,2)
  ax1 = fig.add_subplot(gs[0, 0])

  ax2 = fig.add_subplot(gs[1, 0])
  ax2.sharex = ax1

  ax3 = fig.add_subplot(gs[0, 1])
  ax3.sharey = ax1

  ax4 = fig.add_subplot(gs[1, 1])

  ax5 = fig.add_subplot(gs[2, 1])


  # axe 1 ---------------------------------------------
  ax1.bar(dist_series.index,dist_series, color='b')
  ax1.set_title('Distribution of values')
  ax1.set_ylabel("Nb observations")
  ax1.xaxis.set_tick_params(rotation=45)

  # axe 2 ---------------------------------------------
  sns.histplot(x=dist_series.index, kde=True, stat='density',ax=ax2 )
  ax2.xaxis.set_visible(False)
  ax2.set_title('Density KDE')
  ax2.set_ylabel(colName)
  ax2.invert_yaxis()

  # axe 3 ---------------------------------------------
  sns.histplot(y=dist_series, kde=True, stat='density',ax=ax3 )
  ax3.xaxis.set_visible(False)
  ax3.set_title('Density KDE')
  ax3.set_ylabel("Nb Observation")

  # axe 4 ---------------------------------------------
  sns.boxplot(x=df[colName], ax =ax4,notch=True,color=(1,0,1,0.5))
  ax4.set_xlabel(colName)
  ax4.set_title('Distribution of values Boite a mous')
 

  # axe 5 -----------------------------------------------

  xaxis = np.linspace(0-1/n,1+1/n,len(lorenz)) #one segment of len n for each individual + 1 additional segment at the origine 0. 
  #first segment starts at 0-1/n, las ends at 1+1/n.
  ax5.plot(xaxis,lorenz,drawstyle='steps-post')
  ax5.plot([0,1], [0,1]) #bisector
  ax5.set_title('Lorentz distribution')



  # display ---------------------------------------------

  with graph :
    graph.clear_output()
    plt.show()
  with table :
    table.clear_output()
    full_desc = df[[colName]].describe()
    full_desc.loc["Unique values"] = n 
    full_desc.loc["Missing values%"] = f"{100*df[colName].isna().sum()/len(df[colName]):.2f}"
    full_desc.loc["Kurtosis"] = gamma1
    full_desc.loc["Kurto desc"] = interp_gamma1
    full_desc.loc["Skew"] = gamma2
    full_desc.loc["Skew desc"] = interp_gamma2
    full_desc.loc["Gini"] = gini

    display(full_desc)


  display(hLayout)
```

```python
def cutoffLF(df_values_desc,level=0.01) :

  #let's cut off low frequencies < 1% and keeping top 10 only
  focus_desc1 = df_values_desc[df_values_desc.freq > level ].sort_values("freq")
  lost1 =df_values_desc[df_values_desc.freq < level ]
  display((HTML(f'cutting off low frequency values(<1%), ignoring {len(lost1)} values shown in word cloud : ')))
  #keeping top 10 + others / 
  # maybe it would be better to look for a good trade off between top n and quantile => au filal on fait du PCA artisanal :(
  focus_desc = focus_desc1.head(10)
  lost2 = focus_desc1.iloc[10: , :]
  lost = pd.concat([lost1,lost2])
  focus_desc.loc["other"] = [lost["nb_onservations"].sum(),#nb_observations
              lost["nb_onservations"].sum()/focus_desc.nb_onservations.sum(),#freq
              ]
  focus_desc["cumsum"] = (focus_desc.nb_onservations/focus_desc.nb_onservations.sum()).cumsum()


  lost_asString = ' '.join(lost.index.to_list() )
  if len(lost_asString) >0 :
    wc = WordCloud(width = 800, height = 300,).generate(lost_asString)

    # plot the WordCloud image                        
    plt.axis('off')
    plt.imshow(wc)
    plt.show() 
  else : 
    display(HTML(f"no value has been removed"))

  return focus_desc
```

### Function check if substring in col

```python
def check(subString,df,colName) :
  """
  search for given substring in given colName 
  returns 2 objects :
  - list of unique values matching the substring, as pandas.core.indexes.numeric.Int64Index 
  - list of indexes matching as numpy.ndarray
  """
  
  match_serie = df[colName].str.contains(subString,case=False) # not case sensitive
  matchingRows_list = match_serie[match_serie == True].index
  print(f'{len(matchingRows_list)} individuals match {subString}')
  uv = df.iloc[matchingRows_list][colName].unique()
  uv_counts = df.iloc[matchingRows_list][colName].value_counts()
  print(f"""{len(uv)} values to hamonise :
  {uv_counts}
  """)
  return df.iloc[matchingRows_list][colName].unique(),matchingRows_list
```

### Function describe dataset

```python
def describeDataset(data_location="data/*csv" ) : 
  files = glob.glob(data_location)
  df_list ={}
  for file in files : 
      df_list[file.split("/")[1]] = pd.read_csv(file)
  df_list.keys()

  metaData = pd.DataFrame(columns=["table_name","nb_rows","nb_cols"])

  for df in df_list : 
    #print(f"{df} has {df_list[df].shape[0]} rows and {df_list[df].shape[1]} col")
    to_append = {"table_name" : df,
                      "nb_rows": df_list[df].shape[0],
                      "nb_cols":df_list[df].shape[1],
                      "fillRate": f"{100*df_list[df].count().sum() / df_list[df].size :.1f}%",
                      "duplicated":f"{100*df_list[df].duplicated().sum() / df_list[df].size:.1f}%"
                      }
    metaData = metaData.append( to_append,ignore_index=True )
    
  return df_list,metaData
```
